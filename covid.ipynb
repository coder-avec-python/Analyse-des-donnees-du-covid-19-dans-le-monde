```python
# On supprime qu'on aura pas besoin dans notre analye

data_reduit = data_utile.drop(['Tot Cases/ 1M pop', 'Deaths/ 1M pop', 'Tests/ 1M pop', 'Active Cases'], axis=1)
```

```python
data_reduit
```

```python
sty = go.FigureWidget(layout=go.Layout())
def color(x):
    r='background-color:red'
    y='background-color:purple'
    g='background-color:blue'
    df1 = pd.DataFrame('', index=x.index, colums=x.colums)
    df1.iloc[:,2] = y
    df1.iloc[:,3] = r
    df1.iloc[:,4] = g
    return df1

def show_cases(n):
    n=int(n)
    return data_reduit.sort_values('Total Cases', ascending=False).head(n).style.apply(color, axis=None)

interact(show_cases, n='10')

ipywLayout= widgets.Layout(border= 'solid 2px green')
ipywLayout.display = 'none'
widgets.VBox([sty], layout=ipywLayout)
```
